### Problem 1:

```python
import pandas as pd
def di(AppData):
    df = pd.read_csv(AppData)
    Genre=[]
    Unique_genre=[]
    Dictionary={}
    for i in df.values:
        Genre.append(i[12]) 
    Total_apps=len(Genre)
    Unique_genre=list(set(Genre))
    for j in range(0,len(Unique_genre)):
        a=0
        for i in range(0,len(Genre)):
            if df.iat[i,12]==Unique_genre[j]:
                a=a+1
        percentage=(a/Total_apps)*100
        Dictionary[str(Unique_genre[j])]=percentage
    print(Dictionary)
di('DataFiles/AppleStore.csv')
```

### Problem 3:

```python
def download(AppData):
    used_less, used_moderately, mostly_used, used_more, minimal_usage = [], [], [], [], []
    df = pd.read_csv(AppData)
    for i in df.values:
        if(i[6] > 1000):
            used_less.append(i[2])
        if(i[6] > 10000):
            used_moderately.append(i[2])
            
        if(i[6] > 500000):
            mostly_used.append(i[2])
        
        if(i[6] > 1000000):
            used_more.append(i[2])
            
        if(i[6] < 1000):
            minimal_usage.append(i[2])
    print('used_less Apps:', end = '\n\n')
    print(used_less, end = '\n\n')
    print('used_moderately Apps : ', end = '\n\n')
    print(used_moderately, end = '\n\n')
    print('Expensive Apps : ', end = '\n\n')
    print(mostly_used, end = '\n\n')
    print('Very Expensive Apps : ', end = '\n\n')
    print(used_more, end = '\n\n')
    print('minimal_usage Apps : ', end = '\n\n')
    print(minimal_usage, end = '\n\n')
download('DataFiles/AppleStore.csv')
```
